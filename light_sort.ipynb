```python
import random

def light_sort(list):
    if len(list) <= 1:
        return list
    
    # divide the list into two halves
    mid = len(list) // 2
    left_half = list[:mid]
    right_half = list[mid:]
    
    # recursively sort the two halves
    left_half_sorted = light_sort(left_half)
    right_half_sorted = light_sort(right_half)
    
    # merge the two sorted halves
    return merge_lists(left_half_sorted, right_half_sorted)

def merge_lists(list1, list2):
    result = []
    i = 0
    j = 0

    # iterate over the two lists
    while i < len(list1) and j < len(list2):
        # if the current element in list1 is smaller, increment i
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        # if the current element in list2 is smaller, increment j
        else:
            result.append(list2[j])
            j += 1

    # append any remaining elements from list1
    while i < len(list1):
        result.append(list1[i])
        i += 1

    # append any remaining elements from list2
    while j < len(list2):
        result.append(list2[j])
        j += 1

    return result


```

```python


import random
import time

def quick_sort(list):
    if len(list) <= 1:
        return list
    
    # choose pivot and create two sublists
    pivot = list[0]
    left = []
    right = []
    for i in range(1, len(list)):
        if list[i] < pivot:
            left.append(list[i])
        else:
            right.append(list[i])
    
    # recursively sort the sublists and combine
    return quick_sort(left) + [pivot] + quick_sort(right)

# generate a random list of integers
list = [random.randint(0, 1000) for i in range(1000000)]

# test MergeSort
start_time = time.time()
light_sorted_list = light_sort(list)
light_sort_time = time.time() - start_time

# test QuickSort
start_time = time.time()
quick_sorted_list = quick_sort(list)
quick_sort_time = time.time() - start_time

# compare results and time taken
print("LightSort:")

print("Time Taken: ", light_sort_time)

print("QuickSort:")

print("Time Taken: ", quick_sort_time)

print("Is the result of LightSort equal to the result of QuickSort? ", light_sorted_list == quick_sorted_list)

```
